```python
#accuracy
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

k=range(1,20)
total_acc=[]

for i in k:
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x,y)
  acc=knn.score(x,y)
  total_acc.append(acc)
```

```python
plt.plot(k,total_acc, marker=".")
plt.xticks(np.arange(0,21,1))
plt.show()
```
