```python
x=np.arange(0,10*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
```

```python
a=[1,2,3,4,5]
b=[5,4,6,2,3]
c=[1,4,8,4,2]
plt.plot(a,c,c='#4f7a4b', linestyle='-', linewidth=2, marker='*', label='Line 1')
plt.plot(a,b,c='orange', linestyle='dashed',marker=".", label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Line Graphs')
plt.show()
# plt.plot(b,c)
```

```python
name= ['A','B','C','D','E']
weight=[35,50,55,60,30]
colors=['r','g','b','c','m']
plt.bar(name, weight, color=colors)
plt.xlabel('Names of people')
plt.ylabel("Weight")
plt.show()
```

```python
name= ['A','B','C','D','E']
weight=[35,50,55,60,30]
colors=['r','g','b','c','m']
plt.barh(name, weight, color=colors)
plt.xlabel('Names of people')
plt.ylabel("Weight")
plt.show()
```

```python
#pie chart

hours=['10','8','2','4'] 
work=['sleep','college','food','gym/phone']
col1=['r','g','b','c']
plt.pie(hours, labels=work,colors=col1, startangle=90, explode=[0,0,0.2,0])
plt.legend(title='ROUTUNE')
plt.show()
```
